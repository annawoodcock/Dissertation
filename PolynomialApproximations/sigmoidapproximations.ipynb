```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_approx_3(x):
    return 0.5 + 0.197*x - 0.004*x**3


x = np.linspace(-10, 10, 200)  

plt.plot(x, sigmoid(x), label ='σ(x)')
plt.plot(x, sigmoid_approx_3(x), label='σ3(x)')
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('σ(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def sigmoid_approx_1(x):
    return 0.5 + 0.125*x

plt.plot(x, sigmoid(x), label ='σ(x)')
plt.plot(x, sigmoid_approx_1(x), label='σ1(x)')
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('σ(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python

def sigmoid_approx_5(x):
    return 0.5 + 0.191*x - 0.0046*x**3 + 0.00004*x**5

plt.plot(x, sigmoid(x), label ='σ(x)')
plt.plot(x, sigmoid_approx_5(x), label='σ5(x)')
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('σ(x)')
plt.legend()
plt.grid(True)
plt.show()
```
