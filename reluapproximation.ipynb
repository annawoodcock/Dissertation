```python
import numpy as np
import matplotlib.pyplot as plt

def relu(x):
    return np.maximum(0, x)

def relu_approx(x):
    return 0.375373 + 0.5*x + 0.117071*x**2


x = np.linspace(-6, 6, 400)

plt.figure(figsize=(8, 5))
plt.plot(x, relu(x), label='ReLU(x)')
plt.plot(x, relu_approx(x), label='ReLU(x) approximation')
plt.title('ReLU Activation Function')
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.grid(True)
plt.legend()
plt.show()
```
